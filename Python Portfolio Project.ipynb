```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set_theme(font = 'Century Schoolbook')
mpg = sns.load_dataset('mpg')
```

# Create a visual showing the relationship between horsepower and fuel economy.

```python
sns.relplot (data = mpg, x = 'horsepower', y = 'mpg', aspect = 1.5)
plt.title('Relationship Between Horsepower and Fuel Economy')
plt.xlabel('Horsepower')
plt.ylabel('Miles Per Gallon')
plt.style.use('seaborn-v0_8')
```

```python
# There appears to be a common trend of a negative relationship between horsepower and fuel economy, where cars with higher horsepower generally tend to have worse fuel efficiency.
```

# Create a table summarizing fuel economy by country of origin.

```python
mpg.groupby(['origin']).agg({'mpg':'mean'}).round(1).sort_values(by = 'mpg', ascending = False).reset_index().rename(columns = {'origin':'Country of Origin','mpg':'Average Miles Per Gallon'}).replace({'japan':'Japan','europe':'Europe','usa':'USA'})
```

```python
# Japan model cars have the best average fuel economy while USA has the worst on average.
```

# Create a visual showing how fuel economy has changed over time.

```python
sns.relplot(data = mpg, x = 'model_year', y = 'mpg', kind = 'line',aspect = 1.5, errorbar = None)
plt.title('Change in Fuel Economy Over Time')
plt.xlabel('Model Year')
plt.ylabel('Average Miles Per Gallon')
```

```python
# "The average fuel economy showed a marked improvement between the 1979 and 1980 models.
```

# Create two other visuals or tables related to fuel economy.

```python
mpg['Brand Name'] = mpg['name'].str.split().str[0].str.title()
corrections = {'Vw' : 'Volkswagen','Vokswagen':	'Volkswagen', 'Maxda':'Mazda','Toyouta': 'Toyota','Chevroelt': 'Chevrolet','Amc':'AMC','Bmw':'BMW','Hi':'International Harvester'}
mpg['Brand Name'] = mpg['Brand Name'].replace(corrections)

mpg_brands = mpg.groupby(['Brand Name']).agg({'mpg':'mean'}).round(1).reset_index().rename(columns = {'mpg':'Average Miles Per Gallon'}).sort_values(by = 'Average Miles Per Gallon',ascending = False)
mpg_brands
```

```python
mpg_brands.groupby(['Brand Name']).agg({'Average Miles Per Gallon':'mean'}).round(1).reset_index().sort_values(by = 'Average Miles Per Gallon',ascending = False)
sns.catplot(data = mpg_brands, y = 'Brand Name', x = 'Average Miles Per Gallon', kind = 'bar', height = 7, edgecolor = 'black')
plt.title('Average Miles Per Gallon by Brand Name')
plt.ylabel(None)
plt.xlabel(None)
```

```python
# Between the years 1970 and 1982, the car company that was making the most fuel efficienct cars on average was Nissan.
```
